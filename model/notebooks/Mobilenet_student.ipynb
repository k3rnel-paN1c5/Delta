### Class for Depth Model

```python
class DepthModel(nn.Module):
    """
    Wrapper class for the Depth-Anything model.
    """
    def __init__(self, mod):
        super().__init__()
        self.model = mod
        self.eval()

    def forward(self, x):
        outputs = self.model(x)
        predicted_depth = outputs.predicted_depth
        predicted_depth = F.interpolate(predicted_depth.unsqueeze(1), size=(384, 384), mode='bilinear', align_corners=False)
        return predicted_depth
```

### Class for Student Model

```python
class StudentModel(nn.Module):
    """
    Student model with MobileNetV3 encoder and a custom decoder.
    Designed for real-time inference on edge devices.
    """
    def __init__(self):
        super(StudentModel, self).__init__()
        # mobilenet = mobilenet_v3_large(wieghts=MobileNet_V3_Large_Weights.IMAGENET1K_V1)
        mobilenet = mobilenet_v3_large(pretrained=True)

        self.encoder = mobilenet

         # Freeze encoder parameters
        for param in self.encoder.parameters():
            param.requires_grad = False

        self.decoder_conv_initial = nn.Sequential(
            nn.Conv2d(960, 512, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Dropout(0.25) # Retaining dropout as in your original
        )
        # Upsampling Stage 1: From 12x12 (512ch) to 24x24. Concatenate with encoder's 24x24 features (160ch).
        self.upconv1 = nn.ConvTranspose2d(512, 256, kernel_size=2, stride=2)
        self.decoder_block1 = nn.Sequential(
            nn.Conv2d(256 + 112, 256, kernel_size=3, padding=1), # Sum of upconv output + skip channels (Corrected from 160 to 112)
            nn.ReLU(inplace=True),
            nn.Dropout(0.25) # Retaining dropout as in your original
        )

        # Upsampling Stage 2: From 24x24 (256ch) to 48x48. Concatenate with encoder's 48x48 features (40ch).
        self.upconv2 = nn.ConvTranspose2d(256, 128, kernel_size=2, stride=2)
        self.decoder_block2 = nn.Sequential(
            nn.Conv2d(128 + 40, 128, kernel_size=3, padding=1), # Sum of upconv output + skip channels
            nn.ReLU(inplace=True)
        )

        # Upsampling Stage 3: From 48x48 (128ch) to 96x96. Concatenate with encoder's 96x96 features (24ch).
        self.upconv3 = nn.ConvTranspose2d(128, 64, kernel_size=2, stride=2)
        self.decoder_block3 = nn.Sequential(
            nn.Conv2d(64 + 24, 64, kernel_size=3, padding=1), # Sum of upconv output + skip channels
            nn.ReLU(inplace=True)
        )

        # Upsampling Stage 4: From 96x96 (64ch) to 192x192. Concatenate with encoder's 192x192 features (16ch).
        self.upconv4 = nn.ConvTranspose2d(64, 32, kernel_size=2, stride=2)
        self.decoder_block4 = nn.Sequential(
            nn.Conv2d(32 + 16, 32, kernel_size=3, padding=1), # Sum of upconv output + skip channels
            nn.ReLU(inplace=True)
        )

        # Upsampling Stage 5: From 192x192 (32ch) to 384x384 (final resolution before output conv).
        self.upconv5 = nn.ConvTranspose2d(32, 16, kernel_size=2, stride=2) # Reducing channels before final output

        # Final Output Convolution
        self.output_conv = nn.Conv2d(16, 1, kernel_size=3, padding=1)

    def forward(self, x):
        encoder_features = {}
        for i, module in enumerate(self.encoder.features):
            x = module(x)
            if i == 0:
                encoder_features['192x192'] = x
            elif i == 2:
                encoder_features['96x96'] = x
            elif i == 4:
                encoder_features['48x48'] = x
            elif i == 12:
                encoder_features['24x24'] = x
            elif i == 16:
                encoder_features['12x12'] = x

        # Decoder path
        x = encoder_features['12x12'] # Initial input: 12x12x960
        #print(f"Initial decoder input shape: {x.shape}")

        x = self.decoder_conv_initial(x) # Output: 12x12x512
        #print(f"After decoder_conv_initial shape: {x.shape}")

        x = self.upconv1(x)               # Output: 24x24x256
        #print(f"After upconv1 shape: {x.shape}")
        #print(f"Encoder features '24x24' shape: {encoder_features['24x24'].shape}")
        x = torch.cat([x, encoder_features['24x24']], dim=1) # Concatenate with 24x24x112 skip
        #print(f"After concat with 24x24 features shape: {x.shape}")

        x = self.decoder_block1(x)        # Output: 24x24x256 (after conv)
        #print(f"After decoder_block1 shape: {x.shape}")
        x = self.upconv2(x)               # Output: 48x48x128
        #print(f"After upconv2 shape: {x.shape}")
        #print(f"Encoder features '48x48' shape: {encoder_features['48x48'].shape}")
        x = torch.cat([x, encoder_features['48x48']], dim=1) # Concatenate with 48x48x40 skip
        #print(f"After concat with 48x48 features shape: {x.shape}")

        x = self.decoder_block2(x)        # Output: 48x48x128 (after conv)
        #print(f"After decoder_block2 shape: {x.shape}")
        x = self.upconv3(x)               # Output: 96x96x64
        #print(f"After upconv3 shape: {x.shape}")
        #print(f"Encoder features '96x96' shape: {encoder_features['96x96'].shape}")
        x = torch.cat([x, encoder_features['96x96']], dim=1) # Concatenate with 96x96x24 skip
        #print(f"After concat with 96x96 features shape: {x.shape}")

        x = self.decoder_block3(x)        # Output: 96x96x64 (after conv)
        #print(f"After decoder_block3 shape: {x.shape}")
        x = self.upconv4(x)               # Output: 192x192x32
        #print(f"After upconv4 shape: {x.shape}")
        #print(f"Encoder features '192x192' shape: {encoder_features['192x192'].shape}")
        x = torch.cat([x, encoder_features['192x192']], dim=1) # Concatenate with 192x192x16 skip
        #print(f"After concat with 192x192 features shape: {x.shape}")

        x = self.decoder_block4(x)        # Output: 192x192x32 (after conv)
        #print(f"After decoder_block4 shape: {x.shape}")
        x = self.upconv5(x)               # Output: 384x384x16
        #print(f"After upconv5 shape: {x.shape}")


        x = self.output_conv(x)
        #print(f"Final output shape: {x.shape}")
        return x
```

```python
class StudentModel(nn.Module):
    """
    Student model with MobileNetV3 encoder and a custom decoder.
    Designed for real-time inference on edge devices.
    """
    def __init__(self):
        super(StudentModel, self).__init__()

        deeplabv3 = torchvision.models.segmentation.mobilenet_v3_large(
            weights=MobileNet_V3_Large_Weights.DEFAULT
        )

        self.encoder = deeplabv3.backbone

        # Freeze encoder parameters
        for param in self.encoder.parameters():
            param.requires_grad = False

        # DeepLabV3's ResNet101 backbone typically outputs from layer4 (res5 block)
        # with 2048 channels. We'll design a decoder to upsample this.
        # The original DeepLabV3 has an ASPP module and then upsamples.
        # For simplicity and to match the spirit of the previous decoder (upsampling and convolutions),
        # we'll build a decoder that upsamples iteratively.

        # Initial convolution to reduce channels from 2048
        self.decoder_conv_initial = nn.Sequential(
            nn.Conv2d(2048, 1024, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Dropout(0.25)
        )

        # Upsampling Stage 1: From 24x24 (approx, depends on input size to backbone) to 48x48
        # Assuming an input size of 384x384, the output of ResNet layer4 is 1/16th, so 24x24.
        self.upconv1 = nn.ConvTranspose2d(1024, 512, kernel_size=2, stride=2)
        self.decoder_block1 = nn.Sequential(
            nn.Conv2d(512, 512, kernel_size=3, padding=1), # No skip connection from encoder in this simplified version
            nn.ReLU(inplace=True),
            nn.Dropout(0.25)
        )

        # Upsampling Stage 2: From 48x48 to 96x96
        self.upconv2 = nn.ConvTranspose2d(512, 256, kernel_size=2, stride=2)
        self.decoder_block2 = nn.Sequential(
            nn.Conv2d(256, 256, kernel_size=3, padding=1),
            nn.ReLU(inplace=True)
        )

        # Upsampling Stage 3: From 96x96 to 192x192
        self.upconv3 = nn.ConvTranspose2d(256, 128, kernel_size=2, stride=2)
        self.decoder_block3 = nn.Sequential(
            nn.Conv2d(128, 128, kernel_size=3, padding=1),
            nn.ReLU(inplace=True)
        )

        # Upsampling Stage 4: From 192x192 to 384x384
        self.upconv4 = nn.ConvTranspose2d(128, 64, kernel_size=2, stride=2)
        self.decoder_block4 = nn.Sequential(
            nn.Conv2d(64, 64, kernel_size=3, padding=1),
            nn.ReLU(inplace=True)
        )

        # Upsampling Stage 5: From 384x384 to 384x384 (final resolution before output conv)
        # This is a bit redundant if input size is 384, but useful for consistency or if
        # DeepLab's output resolution is slightly different.
        self.upconv5 = nn.ConvTranspose2d(64, 32, kernel_size=2, stride=2)


        # Final Output Convolution
        self.output_conv = nn.Conv2d(32, 1, kernel_size=3, padding=1)

    def forward(self, x):
        # Encoder pass
        # DeepLabV3's backbone outputs a dictionary of features. We're interested in the final one.
        # The key for the final feature map depends on the backbone. For ResNet, it's typically 'out'.
        encoder_outputs = self.encoder(x)
        x = encoder_outputs['out'] # Assuming the 'out' key for the final feature map

        # Decoder path
        x = self.decoder_conv_initial(x)

        x = self.upconv1(x)
        x = self.decoder_block1(x)

        x = self.upconv2(x)
        x = self.decoder_block2(x)

        x = self.upconv3(x)
        x = self.decoder_block3(x)

        x = self.upconv4(x)
        x = self.decoder_block4(x)

        x = self.upconv5(x)


        # Resize to original input size if necessary (DeepLabV3 sometimes has smaller output)
        # The output of DeepLabV3 backbone might not be exactly 1/16th if there's an ASPP or classification head
        # before this decoder. Here we assume we directly use backbone output.
        # Given the upsampling layers, we ensure it matches the teacher's output size (384, 384).
        x = F.interpolate(x, size=(384, 384), mode='bilinear', align_corners=False)

        x = self.output_conv(x)

        return x
```

### Class for Dataset Loading & Preprocessing

```python
class UnlabeledImageDataset(Dataset):
    """
    Custom dataset for unlabeled images.
    """
    def __init__(self, root_dir, transform=None):
        self.root_dir = root_dir
        self.transform = transform
        self.image_paths = [os.path.join(root_dir, f) for f in os.listdir(root_dir) if f.endswith(('png', 'jpg', 'JPG'))]
        print(f"Found {len(self.image_paths)} images in {root_dir}")

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        img_path = self.image_paths[idx]
        image = Image.open(img_path).convert('RGB')

        if self.transform:
            image = self.transform(image)
        return image
```

### Distillation Loss

```python
class DepthCombinedDistillationLoss(nn.Module):
    """
    Custom loss function for depth distillation.
    Using Pixel wise MSE Loss, SSIM, Gradient MAE Loss, and Scale-Invariant MSE Loss.
    """
    def __init__(self, lambda_depth=1.0, lambda_si=1.0, lambda_grad=1.0, lambda_ssim=1.0, window_size=11):
        super().__init__()
        self.lambda_depth = lambda_depth  # Weight for depth map MSE loss
        self.lambda_si = lambda_si        # Weight for Scale-Invariant MSE loss
        self.lambda_grad = lambda_grad    # Weight for Gradient loss
        self.lambda_ssim = lambda_ssim    # Weight for SSIM loss

        self.mse_depth_loss = nn.MSELoss()  # Mean Squared Error for depth maps
        self.l1_loss = nn.L1Loss()          # L1 Loss for gradients

        # Initialize the SSIM calculation from torchmetrics
        if self.lambda_ssim > 0:
            self.ssim = StructuralSimilarityIndexMeasure(data_range=1.0, kernel_size=window_size)

    def forward(self, student_outputs, teacher_outputs):
        student_depth = student_outputs
        teacher_depth = teacher_outputs

        # Ensure tensors have a batch dimension
        if student_depth.dim() == 3:
            student_depth = student_depth.unsqueeze(0)
        if teacher_depth.dim() == 3:
            teacher_depth = teacher_depth.unsqueeze(0)

        # Ensure tensors have a channel dimension for SSIM
        if student_depth.dim() == 3: # Assuming (Batch, Height, Width)
            student_depth = student_depth.unsqueeze(1) # Becomes (Batch, Channel, Height, Width)
        if teacher_depth.dim() == 3: # Assuming (Batch, Height, Width)
            teacher_depth = teacher_depth.unsqueeze(1) # Becomes (Batch, Channel, Height, Width)


        total_loss = torch.tensor(0.0, device=student_depth.device)

        # 1. MSE Depth Loss
        if self.lambda_depth > 0:
            loss_depth = self.mse_depth_loss(student_depth, teacher_depth)
            total_loss += self.lambda_depth * loss_depth

        # 2. Scale-Invariant MSE Loss
        if self.lambda_si > 0:
            diff = student_depth - teacher_depth
            loss_si = torch.mean(diff**2) - torch.mean(diff)**2
            total_loss += self.lambda_si * loss_si

        # 3. Gradient Loss (using L1 on gradients)
        if self.lambda_grad > 0:
            student_grad_x = torch.abs(student_depth[:, :, :, :-1] - student_depth[:, :, :, 1:])
            student_grad_y = torch.abs(student_depth[:, :, :-1, :] - student_depth[:, :, 1:, :])
            teacher_grad_x = torch.abs(teacher_depth[:, :, :, :-1] - teacher_depth[:, :, :, 1:])
            teacher_grad_y = torch.abs(teacher_depth[:, :, :-1, :] - teacher_depth[:, :, 1:, :])

            loss_grad = self.l1_loss(student_grad_x, teacher_grad_x) + self.l1_loss(student_grad_y, teacher_grad_y)
            total_loss += self.lambda_grad * loss_grad

        # 4. SSIM Loss
        if self.lambda_ssim > 0:
            # Move ssim module to the same device as the tensors
            self.ssim.to(student_depth.device)

            # The torchmetrics SSIM implementation returns a value between -1 and 1.
            # A value of 1 indicates perfect similarity.
            # To use it as a loss, we subtract it from 1.
            d_ssim = self.ssim(student_depth, teacher_depth)
            loss_ssim = (1 - d_ssim) / 2 # Normalize to be between 0 and 1
            total_loss += self.lambda_ssim * loss_ssim

        return total_loss
```

### The Training Function

```python
def train_knowledge_distillation(teacher, student, dataloader, criterion, optimizer, epochs, device):
    """
    Train the student model using Response-Based knowledge distillation.
    """
    teacher.eval() # Teacher should always be in evaluation mode
    student.train() # Student in training mode

    print(f"Starting Knowledge Distillation Training on {device}...")
    models = []
    x = 0
    for epoch in range(epochs):
        running_loss = 0.0
        start_time = time.time()

        for batch_idx, inputs in enumerate(dataloader):
            inputs = inputs.to(device)

            optimizer.zero_grad()

            # Forward pass with Teacher model (no_grad as teacher is fixed)
            with torch.no_grad():
                teacher_outputs = teacher(inputs) # Returns depth map

            # Forward pass with Student model
            student_outputs = student(inputs) # Returns depth map

            # Calculate distillation loss
            loss = criterion(student_outputs, teacher_outputs)

            loss.backward()
            optimizer.step()

            running_loss += loss.item()

            if (batch_idx + 1) % 50 == 0:
                print(f"Epoch [{epoch+1}/{epochs}], Batch [{batch_idx+1}/{len(dataloader)}], Loss: {running_loss / (batch_idx+1):.4f}")

        epoch_loss = running_loss / len(dataloader)
        end_time = time.time()
        print(f"Epoch {epoch+1} finished. Avg Loss: {epoch_loss:.4f}, Time: {end_time - start_time:.2f}s")

        # Save student model checkpoint periodically
        if(epoch+1)%5 == 0:
          models.append(student)
          # torch.save(student.state_dict(), f"/content/drive/MyDrive/distill_any_depth_student_epoch_{epoch+1}.pth")
          # print(f"Student model saved to /content/drive/MyDrive/distill_any_depth_student_epoch_{epoch+1}.pth")

    print("Knowledge Distillation Training Finished!")
    return models
```

# Training Process

### Define Parameters & Models

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Load teacher model
teacher = AutoModelForDepthEstimation.from_pretrained("depth-anything/Depth-Anything-V2-Large-hf").to(device)


teacher_model = DepthModel(teacher).to(device)

# student  model architecture
student_model = StudentModel().to(device)

# Initialize optimizer for the student model
student_optimizer = optim.Adam(student_model.parameters(), lr=1e-4)

# Training parameters
num_epochs = 40

# Distillation Loss
# Instantiate the custom loss function
distillation_criterion = DepthCombinedDistillationLoss(lambda_depth=0.5,  lambda_si=1.0, lambda_grad=1.0, lambda_ssim=1.0, window_size=11)

# transformations for input images (teacher and student will use the same)

input_size = (384, 384)

transform = transforms.Compose([
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomRotation(degrees=10),
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.1),
    transforms.RandomResizedCrop(input_size, scale=(0.8, 1.0)),
    transforms.Resize(input_size),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]) # ImageNet normalization
])

eval_transform = transforms.Compose([
    transforms.Resize(input_size),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# Specify the path to your unlabeled data in Google Drive
unlabeled_data_path = '/content/drive/MyDrive/images/'

# Create dataset and data loader
unlabeled_dataset = UnlabeledImageDataset(root_dir=unlabeled_data_path, transform=transform)
unlabeled_dataloader = DataLoader(unlabeled_dataset, batch_size=5, shuffle=True, num_workers=2)

```
